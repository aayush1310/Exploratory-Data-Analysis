### Univariate Analysis

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']
df_virginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.figure(figsize=(10,8))
plt.plot(df_setosa['SepalLengthCm'], np.zeros_like(df_setosa['SepalLengthCm']),'o')
plt.plot(df_virginica['SepalLengthCm'], np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'], np.zeros_like(df_versicolor['SepalLengthCm']),'o')
plt.xlabel("Length of Sepal")
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df, size=6, col='Species').map(sns.scatterplot, 'SepalLengthCm', 'SepalWidthCm')
```

```python
sns.FacetGrid(df, size=6, hue='Species').map(sns.scatterplot, 'SepalLengthCm', 'SepalWidthCm').add_legend()
```

```python
sns.FacetGrid(df, size=6, col='Species').map(sns.scatterplot, 'PetalLengthCm', 'PetalWidthCm')
```

```python
sns.FacetGrid(df, size=5, hue='Species').map(sns.scatterplot, 'PetalLengthCm', 'PetalWidthCm').add_legend()
```
